```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import KNNImputer
from sklearn.preprocessing import StandardScaler
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='vlag', center=0)
plt.title('Initial Correlation Heatmap')
plt.tight_layout()
plt.show()
```

```python
def replace_zeros_with_null(df, columns):
    df_processed = df.copy()
    for col in columns:
        df_processed[col] = df_processed[col].replace(0, np.nan)
    return df_processed

columns_to_process = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df = replace_zeros_with_null(df, columns_to_process)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(method='pearson', min_periods=1),
            annot=True,
            cmap='vlag',
            center=0)
plt.title('Correlation Heatmap (Pairwise Deletion)')
plt.tight_layout()
plt.show()
```

```python
new_df = df.drop(columns=["SkinThickness","Insulin"])
new_df.head()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(new_df.corr(method='pearson', min_periods=1),
            annot=True,
            cmap='vlag',
            center=0)
plt.title('Correlation Heatmap (Pairwise Deletion)')
plt.tight_layout()
plt.show()
```

```python
new_df.hist(bins = 30, figsize = (10,10), color = 'r')
```

```python
#Imputing the median
new_df['Glucose'].fillna(new_df['Glucose'].median(), inplace = True)
new_df['BloodPressure'].fillna(new_df['BloodPressure'].median(), inplace = True)
new_df['BMI'].fillna(new_df['BMI'].median(), inplace = True)
```

```python
new_df.info()
```

```python
#Train test split
X = new_df.drop('Outcome',axis=1)
y = new_df['Outcome']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

```python
#Scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#Fitting logistic regression model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 1)
classifier.fit(X_train, y_train)
```

```python
y_predict_train = classifier.predict(X_train)
y_predict_train
```

```python
y_train
```

```python
y_predict_train = classifier.predict(X_train)
y_predict_test = classifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_train, y_predict_train)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train, y_predict_train))
```

```python
y_predict_test = classifier.predict(X_test)
y_predict_test
```

```python
cm = confusion_matrix(y_test, y_predict_test)
sns.heatmap(cm, annot=True, fmt="d")
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predict_test))
```

```python
import pickle

# Save the trained model
filename = "diagnosis_trained_model.pkl"
pickle.dump(classifier, open(filename, "wb"))

# Save the scaler
scaler_filename = "scaler.pkl"
pickle.dump(scaler, open(scaler_filename, "wb"))

print("Model and scaler saved successfully!")
```
